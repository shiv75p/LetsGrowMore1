Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

Creating the dataframe

```python
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names = ['sepal length in cm',
                                                                                                  'sepal width in cm',
                                                                                                    'petal length in cm',
                                                                                                    'petal width in cm',
                                                                                                    'class' ])
```

# Model Training and Evaluation

Dividing the dataset into training and testing

```python
data = df.values
X = data[:,0:4]
Y = data[:,4]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4)
```

Here we used 'Support vector Classifier' which is a ML model used for both regression and classification

```python
svn = SVC()
svn.fit(X_train, y_train)
predictions = svn.predict(X_test)
# Calculate the accuracy
accuracy_score(y_test, predictions)
```

```python
print(classification_report(y_test, predictions))
```
